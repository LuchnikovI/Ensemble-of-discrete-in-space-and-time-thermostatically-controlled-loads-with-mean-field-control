```python
import mf_mdp as mm

import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica'], 'size':24})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
rc('xtick', labelsize=24)
rc('ytick', labelsize=24)


############
#parameters
############
n=30
q=9
dr = 0.005
dalpha = 0.4
max_alpha = 60
max_r = 0.7
set_of_epsilon = np.arange(0.01, 0.1, 0.01)
############
############

coef = set_of_epsilon.shape[0] + 1
y = np.arange(0, max_r, dr)
x = np.arange(0, max_alpha, dalpha)
fig = plt.figure()
fig.set_size_inches(15., 9.)
for a in set_of_epsilon:
    gaps = np.array([])
    alpha = 0.
    for k in range(int(max_alpha/dalpha)):
        r = 0.
        for m in range(int(max_r/dr)):
            model = mm.mf_mdp_model(n = n, q = q, epsilon = a, alpha = alpha, r = r)
            gap = model.gap()
            gaps = np.append(gaps, gap)
            r = r + dr
        alpha = alpha + dalpha
    gaps = gaps.reshape(-1, int(max_r/dr))
    gaps = gaps.T
        
    #plotting
        
    plt.contourf(x, y, coef*np.heaviside(gaps, 0), levels=np.arange(set_of_epsilon.shape[0]) + 1, cmap = 'plasma')
    plt.xlabel(r'$\alpha$')
    plt.ylabel(r'$r$')
    coef = coef - 1
plt.savefig(f'Gap_vs_epsilon.pdf')
plt.show()
```
